```python
# Grouping the DataFrame by 'Pclass' and calculating the mean age for each group
mean_age_by_class = df.groupby('Pclass')['Age'].mean()

# Displaying the mean age for each passenger class
mean_age_by_class
```

```python
# now we want add value to the null values of age columns
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if  pd.isnull(Age):
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 30
        else:
            return 25
    else:
        return Age
```

```python
# Assuming you have defined a function called impute_age elsewhere in your code

# Applying the impute_age function to each row of the DataFrame,
# using the values from both the 'Age' and 'Pclass' columns
df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
# Creating dummy variables for the 'Embarked' column
# Drop_first parameter is set to True to drop the first category to avoid multicollinearity
dummies_embarked = pd.get_dummies(df['Embarked'], drop_first=True).astype(int)

# Displaying the first few rows of the dummy variables DataFrame
dummies_embarked.head().astype(int)
```

```python
# Creating dummy variables for the 'Sex' column
# Drop_first parameter is set to True to drop the first category to avoid multicollinearity
dummies_sex = pd.get_dummies(df['Sex'], drop_first=True).astype(int)

# Displaying the first few rows of the dummy variables DataFrame
dummies_sex.head().astype(int)
```

```python
# Concatenating the original DataFrame 'df' with the dummy variables DataFrames 'dummies_sex' and 'dummies_embarked'
# along the columns (axis=1)
new_df = pd.concat([df, dummies_sex, dummies_embarked], axis=1)

# Displaying the first few rows of the new DataFrame 'new_df'
new_df.head()
```

```python
# Dropping the specified columns ('Name', 'Sex', 'Ticket', 'Embarked') from the DataFrame 'new_df'
# axis=1 indicates that columns are to be dropped
df1 = new_df.drop(['Name', 'Sex', 'Ticket', 'Embarked'], axis=1)

# Displaying the first few rows of the new DataFrame 'df1'
df1.head()
```

```python
# Creating a figure with a size of 10x4 inches
plt.figure(figsize=(10, 4))

# Creating the first subplot (1 row, 2 columns, first subplot)
plt.subplot(1, 2, 1)

# Counting the number of passengers who survived for each ticket class and plotting a bar chart
df.loc[df['Survived'] == 1, 'Pclass'].value_counts().sort_index().plot.bar()
plt.title('Number of passengers according\nto ticket class who survived')  # Adding title to the subplot

# Creating the second subplot (1 row, 2 columns, second subplot)
plt.subplot(1, 2, 2)

# Counting the number of passengers who did not survive for each ticket class and plotting a bar chart
df.loc[df['Survived'] == 0, 'Pclass'].value_counts().sort_index().plot.bar()
plt.title('Number of passengers according\nto ticket class who did not survive')  # Adding title to the subplot

# Displaying the subplots
plt.show()
```

```python
# Creating a figure with a size of 18x6 inches
plt.figure(figsize=(18, 6))

# Creating the first subplot (1 row, 2 columns, first subplot)
plt.subplot(1, 2, 1)

# Defining color palette based on survival status
color_gen = {1: "Blue", 0: "red"}

# Creating a count plot to visualize survival rate based on gender
countP_gender = sns.countplot(x='Sex', hue="Survived", data=df, palette=color_gen, alpha=0.7)
for temp_survival in countP_gender.containers:
    countP_gender.bar_label(temp_survival)  # Adding labels to the bars
plt.xlabel("Gender")  # Adding x-axis label
plt.ylabel("Density")  # Adding y-axis label
plt.title("Gender", fontsize=20)  # Adding title to the subplot
plt.legend(labels=['Not Survived', 'Survived'])  # Adding legend

# Creating the second subplot (1 row, 2 columns, second subplot)
plt.subplot(1, 2, 2)

# Calculating gender counts
gender_counts = df['Sex'].value_counts()
labels = gender_counts.index

# Creating a pie chart to visualize gender distribution
plt.style.use('ggplot')
plt.pie(gender_counts, labels=labels, autopct='%.2f%%', startangle=90)
plt.axis('equal')  # Making sure the pie chart is circular
plt.legend(loc='upper left')  # Adding legend to the pie chart
circle = plt.Circle(xy=(0, 0), radius=0.75, facecolor='white')
plt.gca().add_artist(circle)  # Adding a circle to the pie chart
plt.title('Gender Distribution')  # Adding title to the subplot

# Displaying the subplots
plt.show()
```

```python
# Define colors for survived (1) and not survived (0)
color_em = {1: "Blue", 0: "red"}

# Creating a count plot to visualize survival distribution based on the port of embarkation ('Embarked')
countP_em = sns.countplot(x='Embarked', hue="Survived", data=df, palette=color_em, alpha=0.7)

# Adding labels to the bars in the count plot
for temp_survival in countP_em.containers:
    countP_em.bar_label(temp_survival)

# Adding labels and title to the count plot
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.title("Survival Distribution based on Port of Embarkation", fontsize=20)

# Adding legend to the count plot
plt.legend(labels=['Not Survived', 'Survived'])

# Displaying the count plot
plt.show()
```
